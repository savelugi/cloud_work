```python
from visualization import *
from network_graph import *


#interplayer delay optimizations with variable migration costs
csv_file_name = '/Users/ebenbot/Documents/University/cloud_work/saves/dynamic/20241205110149_usa/20241205110149_usa.csv' #0
csv_file_namee = '/Users/ebenbot/Documents/University/cloud_work/saves/dynamic/20241205110347_usa/20241205110347_usa.csv' #0.4
csv_file_nameee = '/Users/ebenbot/Documents/University/cloud_work/saves/dynamic/20241205110540_usa/20241205110540_usa.csv' #0.7

csvs = [csv_file_name,csv_file_namee,csv_file_nameee]
mod_cols = ['tick', 'model_type', 'num_players', 'nr_of_active_game_servers', 'nr_of_player_migrations', 'total_migration_cost', 'nr_of_server_migrations', 'nr_of_players_moved',
        'average_player_to_server_delay', 'min_player_to_server_delay', 'max_player_to_server_delay', 'average_player_to_player_delay', 'min_player_to_player_delay',
        'max_player_to_player_delay', 'QoE_score', 'sim_time', 'migration_cost']
for mod_col in mod_cols:
    plt.figure(figsize=(10, 6))
    for csv_file_name in csvs:
        df_results = pd.read_csv(csv_file_name, comment='#')
        mod_data = df_results[mod_cols]
        data = mod_col
        sns.lineplot(data=mod_data, x='tick', y=data, label=f'migration_cost = {df_results['migration_cost'][2]}')
    plt.xlabel('Tick')
    plt.ylabel(data)
    plt.title(f'{data} vs tick')
    plt.legend()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd



ydatas = ['nr_of_player_migrations', 'nr_of_server_migrations']
for ydata in ydatas:
    plt.figure(figsize=(10, 6))
    all_data = pd.concat([pd.read_csv(csv, comment='#') for csv in csvs])
    bar_data = all_data.groupby(['tick', 'migration_cost'])[ydata].mean().reset_index()
    bar_data = bar_data[bar_data[ydata] > 0]

    # Barplot generálása
    sns.barplot(data=bar_data, x='tick', y=ydata, hue='migration_cost', palette='coolwarm')

    # Címkék és cím hozzáadása a barplothoz
    plt.xlabel('Tick')
    plt.ylabel(ydata)
    plt.title(f'{ydata} vs tick')
    plt.legend()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

all_data = pd.concat([pd.read_csv(csv, comment='#') for csv in csvs])

plt.figure(figsize=(10, 6))
#sns.boxplot(data=all_data, x='migration_cost', y='max_player_to_player_delay')
sns.violinplot(data=all_data, x='migration_cost', y='max_player_to_player_delay')
plt.xlabel('Migration Cost')
plt.ylabel('Average Player-to-Player Delay')
plt.title('Distribution of Player-to-Player Delay by Migration Cost')
plt.ylim(10, 50)
plt.show()
```
